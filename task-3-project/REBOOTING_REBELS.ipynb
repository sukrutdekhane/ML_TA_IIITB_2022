### 1.4.2 Handle NULL values in columns with <5% or >60% NULL values

```python
index_list = main_df_train.columns.to_list()
index_list = index_list[1:len(index_list)-1]
```

```python
lower_threshold = int(len(main_df_train)*0.05)
upper_threshold = int(len(main_df_train)*0.6)
```

For columns with <5% NULL values, if occurence of mode is greater than 80% in the column, then replace NULL values with mode, otherwise drop the rows in training data.

For columns with >60% NULL values, drop the columns

```python
for index in index_list:
    if(main_df_train[index].isna().sum() < lower_threshold):
        mode_train = main_df_train[index].mode(dropna=True).to_numpy()
        mode_train = mode_train[0]
        mode_test = main_df_test[index].mode(dropna=True).to_numpy()
        mode_test = mode_test[0]
        mode_threshold_train = np.int_(main_df_train[index].describe()['count']*0.8)
        mode_num_train = main_df_train[index].value_counts()[mode_train]
        mode_threshold_test = np.int_(main_df_test[index].describe()['count']*0.8)
        mode_num_test = main_df_test[index].value_counts()[mode_test]
        if(mode_num_train > mode_threshold_train):
            # plot_pie(index)
            main_df_train[index].fillna(mode_train, inplace=True)
            main_df_test[index].fillna(mode_test, inplace=True)
        else:
            main_df_train.drop(axis="rows", labels=main_df_train.index[main_df_train[index].isna()], inplace=True)
    elif(main_df_train[index].isna().sum() > upper_threshold):
        main_df_train.drop(axis="columns", labels=index, inplace=True)
        main_df_test.drop(axis="columns", labels=index, inplace=True)
        
        
```

```python
main_df_train.interpolate(method ='linear', limit_direction ='forward', inplace=True)
main_df_test.interpolate(method ='linear', limit_direction ='forward', inplace=True)
```

For non - categorical columns, if occurence of mode is >80%, then replace NULL values with mode, else replace NULL values with median

For categorical columns, replace NULL values with mode

```python
index_list = main_df_train.columns.to_list()
index_list = index_list[1:len(main_df_train.columns)-1]
for index in index_list:
    mode_train = main_df_train[index].mode(dropna=True).to_numpy()
    mode_train = mode_train[0]
    mode_test = main_df_test[index].mode(dropna=True).to_numpy()
    mode_test = mode_test[0]
   
    mode_threshold_train = np.int_(main_df_train[index].describe()['count']*0.8)
    mode_num_train = main_df_train[index].value_counts()[mode_train]
    mode_threshold_test = np.int_(main_df_test[index].describe()['count']*0.8)
    mode_num_test = main_df_test[index].value_counts()[mode_test]
    if(mode_num_train > mode_threshold_train):
        main_df_train[index].fillna(mode_train, inplace=True)
        main_df_test[index].fillna(mode_test, inplace=True)
   
    else:
        if(main_df_train[index].dtype == np.int64 or main_df_train[index].dtype == np.float64):
            
            median_train = main_df_train[index].median()
            median_test = main_df_test[index].median()
            main_df_train[index].fillna(median_train, inplace=True)
            main_df_test[index].fillna(median_test, inplace=True)
        
        else:
            # plot_pie(index)
            main_df_train[index].fillna(mode_train, inplace=True)
            main_df_test[index].fillna(mode_test, inplace=True)
```

# 2. Training Models

```python
def train_model(model_num, X_train, Y, X_test):
    if(model_num == 1):
        print("Logistic Regression")
        model = LogisticRegression(max_iter=10000)
    elif(model_num == 2):
        print("Naive Bayes")
        model = GaussianNB()
    elif(model_num == 3):
        print("Decision Tree")
        model = DecisionTreeClassifier(random_state = 42)
    elif(model_num == 4):
        print("Random Forest")
        model = RandomForestClassifier(n_estimators=1000)
    elif(model_num == 5):
        print("XGBoost")
        model = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss')
    elif(model_num == 6):
        print("LightGBM")
        model = lgb.LGBMClassifier(n_estimators=1000)
    elif(model_num == 7):
        print("Tuned LightGBM")
        model = lgb.LGBMClassifier(      
        feature_pre_filter = False,
         num_leaves= 512,
         feature_fraction= 0.3,
         min_child_samples= 35,
         learning_rate= 0.009,
         n_estimators= 1000,
         class_weight= 'balanced'
      )
    model.fit(X_train, Y)
    train_ypred = model.predict(X_train)
    train_ypred = train_ypred.reshape((train_ypred.shape[0], 1))
    test_ypred = model.predict(X_test)
    test_ypred = test_ypred.reshape((test_ypred.shape[0], 1))
    return(train_ypred, test_ypred)
```

We do not have the LGBM classifier hyperparameter values which gave us the best results on Kaggle as we tuned the model manually and private leaderboard was not visible before the deadline. For most of our submissions with kaggle score > 0.60, we just varied our learning rate between 0.0065 and 0.01. The values of remaining hyperparameters were kept as they are in the above code block.

```python
def print_scores(Y, Yhat):
    # print(f'accuracy: {accuracy_score(Y, Yhat)}')
    # print(f'precision: {precision_score(Y, Yhat)}')
    # print(f'recall: {recall_score(Y, Yhat)}')
    print(f'f1: {f1_score(Y, Yhat, average="macro")}')
```
