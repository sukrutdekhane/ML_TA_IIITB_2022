```python
import numpy as np
import pandas as pd 
import os
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

### Importing DataSet

```python
train_df = pd.read_csv('../input/ml-project-final-dataset/processedTraining.csv')
test_df = pd.read_csv('../input/ml-project-final-dataset/processedTesting.csv')
```

### Importing re for renaming columns for LightGBM

```python
import re
train_df = train_df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
test_df = test_df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
test_SKID = test_df['sk_id_curr']
test_features = test_df.drop(['sk_id_curr'], axis=1)
```

```python
main_X = train_df.drop(['target'], axis=1)
main_y = train_df['target']
X_train, X_test, y_train, y_test = train_test_split(main_X, main_y, test_size = 0.3, random_state = 42)
```

LGBMClassifier(boosting_type='gbdt', num_leaves=31, max_depth=-1, learning_rate=0.1, n_estimators=100, subsample_for_bin=200000, objective=None, class_weight=None, min_split_gain=0.0, min_child_weight=0.001, min_child_samples=20, subsample=1.0, subsample_freq=0, colsample_bytree=1.0, reg_alpha=0.0, reg_lambda=0.0, random_state=None, n_jobs=None, importance_type='split', **kwargs)

### Model Implementation

```python
lgbm_model = LGBMClassifier(n_estimators=200,max_depth=19, learning_rate=0.3, class_weight='balanced')
lgbm_model.fit(X_train, y_train)
y_pred = lgbm_model.predict(X_test) 
f1_score(y_test, y_pred, average='macro')
```

### Fitting on whole training data

```python
final_model_grid = LGBMClassifier(n_estimators=200,max_depth=19, learning_rate=0.2, class_weight='balanced')
final_model_grid.fit(main_X, main_y)
test_pred = final_model_grid.predict(test_features)
```

```python
test_df["target"] = test_pred
to_submit = test_df[["sk_id_curr", "target"]]
to_submit.to_csv('LgbmClassifier2.csv', index=False)
```
